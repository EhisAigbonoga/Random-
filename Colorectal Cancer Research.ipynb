```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from statsmodels.stats import weightstats
import scipy.stats as stats
df = pd.read_csv('Colorectal Cancer Research.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Assuming you have a dataframe called 'df' with columns 'column1' and 'column2'
correlation_matrix = df[['KnowTotal', 'RiskTotal']].corr()
```

```python
# Assuming you have a dataframe called 'df' with columns 'column1' and 'column2'
# Replace 'column1' and 'column2' with your actual column names

# Calculate the correlation coefficient between the two columns
correlation_coefficient = df['RiskTotal'].corr(df['KnowTotal'])

# Plot the line graph
plt.plot(df['RiskTotal'], df['KnowTotal'], marker='o', linestyle='-', color='b', label=f'Correlation: {correlation_coefficient:.2f}')
plt.xlabel('RiskTotal')
plt.ylabel('KnowTotal')
plt.title('Line Graph Correlation Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python

correlation_coefficient = df['KnowTotal'].corr(df['RiskTotal'])

# Plot the dots as a scatter plot
plt.scatter(df['RiskTotal'], df['KnowTotal'], marker='o', color='b', label=f'Correlation: {correlation_coefficient:.2f}')
plt.xlabel('Total Risk Score')
plt.ylabel('KnowTotal')
plt.title('Scatter Plot Correlation of Knowledge and Risk')
plt.legend()
plt.grid(True)
plt.show()
```
